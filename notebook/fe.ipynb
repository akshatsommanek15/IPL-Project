```python
df_copy=df_copy[df_copy['innings']==2]
```

```python
drop_cols = [
    'Unnamed: 0', 'date', 'match_type', 'event_name','match_id',
    'bat_pos', 'bowler', 'batter',
    'runs_bowler', 'runs_not_boundary',
    'extra_type', 'non_striker', 'non_striker_pos',
    'wicket_kind', 'player_out', 'fielders',
    'review_batter', 'team_reviewed', 'review_decision',
    'umpire', 'umpires_call', 'player_of_match',
    'day', 'month', 'year', 'season',
    'gender', 'team_type', 'superover_winner',
    'result_type', 'method',
    'event_match_no', 'stage', 'match_number',
    'new_batter', 'batter_runs', 'batter_balls',
    'bowler_wicket', 'batting_partners',
    'next_batter', 'striker_out','win_outcome'
]
```

```python
df_copy.drop(columns=drop_cols,inplace=True)
```

```python
df_copy=df_copy.dropna(subset=['runs_target'])
```

```python
df_copy.head()
```

```python
df_copy['current_score']=df_copy['team_runs']
df_copy['balls_bowled']=df_copy['team_balls']
df_copy['wickets']=df_copy['team_wicket']
```

```python
df_copy['balls_left']=120-df_copy['team_balls']
df_copy['runs_left']=df_copy['runs_target']-df_copy['team_runs']
df_copy['wickets_left']=10-df_copy['team_wicket']
```

```python
df_copy['current_run_rate']=round(df_copy['team_runs']/(df_copy['team_balls']/6),2)
df_copy['required_run_rate']=round(df_copy['runs_left']/(df_copy['balls_left']/6),2)
```

```python
df_copy=df_copy[df_copy['balls_left'] > 0]
```

```python
df_copy.drop(columns=['team_runs', 'team_balls', 'team_wicket'],inplace=True)
```

```python
df_copy.shape
```

```python
df_copy.columns
```

```python
drop_more = [
    'runs_batter', 
    'balls_faced', 
    'valid_ball', 
    'runs_extras', 
    'runs_total'
]

df_copy=df_copy.drop(columns=drop_more)
```

```python
df_copy=df_copy.drop(columns=['balls_per_over', 'overs'])
```

```python
df_copy.shape
```

```python
df_copy.head()
```

```python
df_copy['result']=(df_copy['batting_team']==df_copy['match_won_by']).astype(int)
```

```python
df_copy.drop('match_won_by',axis=1,inplace=True)
```

```python
col_drop=['over', 'ball', 'ball_no','current_score', 'balls_bowled', 'wickets','innings']
```

```python
df_copy.drop(columns=col_drop,inplace=True)
```

```python
df_copy['toss_advantage'] = (df_copy['toss_winner'] == df_copy['batting_team']).astype(int)
```
